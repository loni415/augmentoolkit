```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Period": ["2021H1", "2021H2", "2022H1", "2022H2", "2023H1", "2023H2", "2024H1"],
    "Over 3B RMB": [0, 2, 0, 2, 0, 0, 2],
    "1-3B RMB": [5, 5, 2, 2, 2, 3, 0],
    "100M-1B RMB": [29, 31, 25, 14, 29, 30, 17],
    "50-100M RMB": [9, 18, 19, 16, 4, 5, 4],
    "10-50M RMB": [24, 23, 12, 12, 46, 36, 18],
    "<10M RMB": [4, 17, 21, 13, 3, 7, 4],
    "Not disclosed": [264, 327, 241, 218, 163, 204, 103]
}
df = pd.DataFrame(data)
df.set_index("Period", inplace=True)

# Chart 1: Disclosed categories only
disclosed = df.drop(columns=["Not disclosed"])
fig, ax = plt.subplots(figsize=(12, 6))
disclosed.plot(kind="bar", ax=ax)
plt.title("AI Financing Events by Fund Size (Disclosed Only)")
plt.xlabel("Period")
plt.ylabel("Number of Events")
plt.legend(title="Fund Size", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# Chart 2: Not disclosed only
fig, ax = plt.subplots(figsize=(8, 4))
df["Not disclosed"].plot(kind="bar", color="gray", ax=ax)
plt.title("AI Financing Events: Not Disclosed")
plt.xlabel("Period")
plt.ylabel("Number of Events")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Period": ["2021H1", "2021H2", "2022H1", "2022H2", "2023H1", "2023H2", "2024H1"],
    "Over 3B RMB": [0, 2, 0, 2, 0, 0, 2],
    "1-3B RMB": [5, 5, 2, 2, 2, 3, 0],
    "100M-1B RMB": [29, 31, 25, 14, 29, 30, 17],
    "50-100M RMB": [9, 18, 19, 16, 4, 5, 4],
    "10-50M RMB": [24, 23, 12, 12, 46, 36, 18],
    "<10M RMB": [4, 17, 21, 13, 3, 7, 4],
    "Not disclosed": [264, 327, 241, 218, 163, 204, 103]
}
df = pd.DataFrame(data)
df.set_index("Period", inplace=True)

# Order columns so "Not disclosed" is last (top of stack)
columns_order = [
    "Over 3B RMB", "1-3B RMB", "100M-1B RMB", "50-100M RMB", "10-50M RMB", "<10M RMB", "Not disclosed"
]
df = df[columns_order]

# Colors: muted for "Not disclosed"
colors = [
    "#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", "#8c564b", "#cccccc"
]

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
bottom = None
for i, col in enumerate(df.columns):
    if col == "Not disclosed":
        # Lighter color, dashed edge for "Not disclosed"
        bars = ax.bar(
            df.index, df[col], bottom=bottom, label=col, color=colors[i],
            edgecolor='black', hatch='//', alpha=0.7
        )
    else:
        bars = ax.bar(
            df.index, df[col], bottom=bottom, label=col, color=colors[i]
        )
    if bottom is None:
        bottom = df[col]
    else:
        bottom += df[col]

plt.title("Stacked Bar Chart: AI Financing Events by Fund Size (2021H1-2024H1)")
plt.xlabel("Period")
plt.ylabel("Number of Events")
plt.legend(title="Fund Size", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
